```python
import jarvisBot as pc
from tkinter import *
```

```python
# Função que será chamada ao clicar no botão ou pressionar "Enter"
def roda_Chatbot(event=None):  # 'event=None' permite que a função seja chamada com ou sem evento de teclado
    global entrada_sugestao
    global entrada_nome_usuario
    global nome_usuario

    texto = e_mensagem.get()  # Obtém o texto inserido pelo usuário
    e_mensagem.delete(0, END)  # Limpa o campo de entrada após enviar
    
    if entrada_nome_usuario:  # Se for a primeira vez, solicita o nome do usuário
        nome_usuario = texto
        text_area.insert(END, f"{nome_usuario}: {texto}\n\n")  # Exibe a mensagem do usuário
        saudacao = pc.saudacao_GUI()  # Obtém a saudação do bot
        text_area.insert(END, f"{nome_maquina}: {saudacao}\n")  # Adiciona a saudação com uma nova linha
        text_area.see(END)  # Rolagem automática para o final
        entrada_nome_usuario = False  # Define que o nome já foi fornecido
    else:
        text_area.insert(END, f"{nome_usuario}: {texto}\n")  # Exibe a mensagem do usuário
        text_area.see(END)  # Rolagem automática para o final

        if entrada_sugestao:  # Caso o bot solicite uma sugestão do usuário
            pc.salva_sugestao(texto)  # Salva a sugestão
            entrada_sugestao = False
            resposta = "Agora aprendi! Vamos continuar nossa conversa...\n"
            text_area.insert(END, f"{resposta}\n")  # Adiciona resposta do bot com nova linha
            text_area.see(END)  # Rolagem automática para o final
        else:
            resposta = pc.buscaResposta_GUI("Cliente: " + texto + "\n")  # Busca resposta
            pc.setHistorico(texto)
            if resposta == "Me desculpe, não sei o que falar":  # Caso o bot não saiba o que responder
                resposta = "Me desculpe, não sei o que falar. O que você esperava?"
                entrada_sugestao = True  # Habilita a entrada de sugestão
            else:
                resposta = pc.exibeResposta_GUI(texto, resposta)  # Exibe a resposta apropriada
            
            text_area.insert(END, f"{resposta}\n")  # Adiciona a resposta com nova linha
            text_area.see(END)  # Rolagem automática para o final

# Criação da janela principal
main_window = Tk()
main_window.title("Jarvis")
main_window.geometry("400x400")

# Frame para a entrada de mensagem
frame = Frame(main_window)
frame.grid()

# Rótulo de identificação
l_identif = Label(frame, text="Insira uma mensagem aqui: ")
l_identif.grid(row=0, column=0)

# Campo de entrada de mensagem
e_mensagem = Entry(frame)
e_mensagem.grid(row=0, column=1)

# Botão para enviar a mensagem
Button(frame, text="Clique", command=lambda: roda_Chatbot()).grid(row=0, column=2)

# Associa a tecla "Enter" ao envio da mensagem
e_mensagem.bind("<Return>", roda_Chatbot)  # Quando "Enter" é pressionado, chama roda_Chatbot

# Frame para exibir o histórico de conversa com rolagem
frame2 = Frame(main_window)
frame2.grid(row=1, column=0, columnspan=3)

# Barra de rolagem para o histórico de conversa
scrollbar = Scrollbar(frame2)
scrollbar.grid(row=0, column=1, sticky='ns')

# Área de texto onde será mostrado o histórico de conversas
text_area = Text(frame2, yscrollcommand=scrollbar.set, wrap="word", height=20, width=50)
text_area.grid(row=0, column=0)
scrollbar.config(command=text_area.yview)

# Variáveis iniciais
nome_maquina = "Jarvis"
entrada_sugestao = False
entrada_nome_usuario = True
nome_usuario = ""

# Configuração inicial do bot (chamada de algum método para definir o nome do bot)
pc.setNomeBot(nome_maquina)

# Início da conversa com uma pergunta do bot
text_area.insert(END, f"{nome_maquina}: Qual o seu nome?\n")
text_area.see(END)  # Rolagem automática para o final

# Loop principal da interface gráfica
main_window.mainloop()
```
